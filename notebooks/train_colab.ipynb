# 🎯 YOLO — Transfer Learning no Google Colab

Este notebook realiza o treinamento da rede YOLO com transfer learning.

**Pré-requisito:** Ative a GPU em `Ambiente de execução → Alterar tipo de ambiente de execução → GPU`

```
# 1. Verificar GPU disponível
!nvidia-smi
```

```
# 2. Clonar YOLOv5
!git clone https://github.com/ultralytics/yolov5.git
%cd yolov5
!pip install -r requirements.txt
```

```
# 3. Fazer upload do dataset (ou montar Google Drive)
from google.colab import drive
drive.mount('/content/drive')

# Descompacte seu dataset
# !unzip /content/drive/MyDrive/dataset.zip -d /content/dataset
```

```
# 4. Criar data.yaml
data_yaml = """
path: /content/dataset
train: images/train
val:   images/val
test:  images/test

nc: 2
names:
  0: classe_1
  1: classe_2
"""

with open('/content/data.yaml', 'w') as f:
    f.write(data_yaml)

print('data.yaml criado!')
```

```
# 5. Treinar com transfer learning (pesos pré-treinados no COCO)
!python train.py \
    --data /content/data.yaml \
    --weights yolov5s.pt \
    --epochs 100 \
    --batch-size 16 \
    --img 640 \
    --project /content/results \
    --name exp1 \
    --cache
```

```
# 6. Visualizar métricas de treinamento
from IPython.display import Image
Image('/content/results/exp1/results.png')
```

```
# 7. Inferência nas imagens de teste
!python detect.py \
    --weights /content/results/exp1/weights/best.pt \
    --source /content/dataset/images/test/ \
    --conf-thres 0.4 \
    --project /content/results \
    --name detections
```

```
# 8. Visualizar resultados
import glob
from IPython.display import Image, display

for img_path in glob.glob('/content/results/detections/*.jpg')[:5]:
    display(Image(img_path, width=640))
```
